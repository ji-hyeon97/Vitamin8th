#시각화란?
- 데이터 분석 결과를 시각적으로 보기 쉽도록 정리하는 것
- 데이터에 담긴 의미를 쉽게 찾을 수 있도록 만드는 과정
- 데이터 분석의 효율성을 확보하는 과정

#시각화의 장점
- 유의미한 동향과 인사이트를 신속하게 파악하여 더욱 정확한 의사결정 가능
- 많은 양의 데이터를 한눈에 파악
- 폭넓은 분석방법론 제시
- 데이터 분석의 정확성 향상

#데이터 타입
- 범주형 데이터 1.명목형 데이터(성별, 혈액형) 2. 순서형 데이터(난이도, 평점)

- 수치형 데이터 1.이상형 데이터(판매량, 불량품 수) 2.연속형 데이터(키, 몸무게, 나이)

#그래프 타입
-1차원 1. 범주형(막대그래프, 파이차트) 2. 연속형(히스토그램, 박스플롯, 바이올린)

-2차원 1.히트맵 2.박스플롯 3. 산점도

#막대그래프란?
- 각 범주의 데이터 값의 크기에 비례하여 직사각형 막대로 표현
- 막대 높이의 상대적 길이 차이를 통해 크고 작음을 설명
- 시계열 데이터를 설명하기 적합한 ‘bar’ 
- 각 변수 사이 값의 크기 차이를 설명하기 적합한 ‘barh’ 

```python
import matplotlib.pyplot as plt
import pandas as pd

labels = ['Kor', 'Maths', 'Eng', 'Sci']
scores = [80,90,75,100]

plt.bar(labels,scores) #가로 세로
plt.show()

#plt.bar() : 가로형 막대 그래프
```

```python
plt.barh(labels,scores) #가로 세로
plt.show()

#plt.barh() : 세로형 막대그래프
```

```python
plt.bar(labels, scores, align='center', color='green', alpha=0.5)
#align은 눈금과 막대의 위치를 조절합니다. 
#디폴트 값은 center이며, edge로 설정하면 막대의 왼쪽 끝에 눈금이 표시됩니다.
#alpha는 투명도

plt.xticks(labels, fontsize=15, rotation=0)
plt.yticks(fontsize=15, rotation=0)
#x축, y축 눈금 설정
#x축 라벨을 0도로 만들고 글자 크기 설정
#y축 값을 0도로 만들고 글자 크기 설정

plt.xlabel('Subject', fontsize=15)
plt.xlabel('Scores', fontsize=15)
plt.title('Scores for midterm exam', fontsize=20)
#label 및 title 설정

plt.show()
```

```python
plt.barh(labels, scores, align='center', color='red', alpha=0.5)
#빨강색 지정

plt.xticks(fontsize=15, rotation=30)
plt.yticks(labels, fontsize=15, rotation=30)
#x축, y축 눈금 설정

plt.xlabel('Subject', fontsize=15)
plt.xlabel('Scores', fontsize=15)
plt.title('Scores for midterm exam', fontsize=20)
#label 및 title 설정

plt.show()
```

#파이 차트란?
- 범주형 구성 비율을 원형으로 표현한 그래프
- 부채꼴의 중심각을 구성 비율에 비례하도록 표현

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

labels = ["Dog", "Cat", "Fish","Others"]
sizes=[45,30,15,10]

plt.pie(sizes)
plt.show()
```

```python
plt.pie(sizes,
         labels = labels, # 파이차트 밖에 라벨 표시
         autopct = '%1.1f%%', # 파이차트 안에 수치 표현
         explode=(0.2,0,0,0), # Dog만 튀어나오도록 세팅
         shadow = True, # 그림자
         startangle = 90 # 파이를 그리기 시작하는 위치
         )

plt.title('Animals',fontsize = 15)
plt.show()
```

- explode : 부채꼴이 파이 차트의 중심에서 벗어나는 정도를
설명
- autopct : 부채꼴 안에 표시될 숫자의 형식을 지정
-> %.1f : 소수점 첫째자리까지 보여줌
-> %.1f%% : 소수점 첫째자리까지 보여주고 뒤에 %
-> %d%% : 가장 가까운 정수로 반올림하고 맨 뒤에 %
- startangle : 부채꼴이 그려지는 시작 각도를 설정
- shadow : True로 설정하면 파이 차트에 그림자가 표시됨

```python
ratio=[34,32,16,18]
labels=['Apple','Banana','Melon','Grapes']
explode=[0.05, 0.05, 0.05, 0.05]
colors=['red','yellow','green','purple']
wedgeprops={'width':0.7, 'edgecolor' : 'w', 'linewidth':5}
# width : 넓이 , edgecolor : white, linewidth : 공백선의 넓이

plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260,
        counterclock=False, colors=colors, wedgeprops=wedgeprops)
```

- counterclock : False로 설정하면 시계 방향 순서로 부채꼴
영역이 표시됨
- wedgeprops : 부채꼴 영역의 스타일을 설정

#히스토그램이란?
- 도수분포표를 그래프로 나타낸 것으로 주로 연속형 데이터 분포를 파악할 때 많이
사용
- X축 : 데이터의 전범위를 같은 크기의 여러 구간으로 구분
- Y축 : 각 구간에 속한 데이터의 빈도수 표현


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

N=10000
x=np.random.randn(N)
#평균0, 표준편차1의 가우시안 표준정규분포 난수를 matrix array(m,n) 생성

plt.hist(x,bins=30)
#bins = ->구간 개수 설정
plt.show()
```

```python
plt.hist(x,bins=30, density=True, cumulative=True)
plt.yticks(fontsize=15)

plt.show()
```

- bins : 몇 개의 영역으로 나눌지 설정
- density : True일 때 확률밀도를 형성시키기 위해
정규화, 히스토그램 값들의 면적 값은 1이 됨
- cumulative : 누적분포

```python
plt.hist(x, bins=30, color='green',alpha=0.5)
plt.hist(x+5, bins=60, color='orange', alpha=0.5)

plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

plt.legend(['histogram1', 'histogram2'],fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

spread = np.random.rand(50)*100 #0~100사이 값을 50개 반환
center = np.ones(25)*50 #50로 초기화된 값 25개
flier_high=np.random.rand(10)*100+100 #100~200 사이의 값을 10개 반환
flier_low =np.random.rand(10)*100-100 #-100~0 사이의 값을 10개 반환
data=np.concatenate((spread,center,flier_high,flier_low)) #넘파이 배열 합치기

plt.boxplot(data)
plt.show()

#np.random.rand() : 0~1사이의 값의 균등 분포라고 불리는 값을 반환하는 함수
```

```python
plt.boxplot(data,vert=False)

#vert: False일때 가로
```

```python
spread = np.random.rand(50)*100 #0~100사이 값을 50개 반환
center1 = np.ones(25)*50 #50로 초기화된 값 25개
center2 = np.ones(25)*50 #50로 초기화된 값 25개

flier_high=np.random.rand(10)*100+100 #100~200 사이의 값을 10개 반환
flier_low =np.random.rand(10)*100-100 #-100~0 사이의 값을 10개 반환

data1=np.concatenate((spread,center1,flier_high,flier_low)) #넘파이 배열 합치기
data2=np.concatenate((spread,center2,flier_high,flier_low)) #넘파이 배열 합치기
data=[data1,data2,data[::5]]
#data[::5] ??

plt.boxplot(data)
plt.show()
```

```python
stages =['Website visit','Downloads','Potential customers',
         'Requested price','invoice sent']
df_mtl = pd.DataFrame(dict(number=[39,27.4,20.6,11,3],stage=stages))
df_mtl['office'] = 'Montreal'
df_toronto = pd.DataFrame(dict(number=[52,36,18,14,5],stage=stages))
df_toronto['office']='Toronto'
df=pd.concat([df_mtl,df_toronto],axis=0)
fig=px.funnel(df, x='number',y='stage',color='office')
fig.show()
```

#Funnel chart
비즈니스 프로세스의 다양한 단계에서의 데이터를 나타내기 위해 사용되는 차트
판매 프로세스의 수익, 손실을 관찰하고 점차적으로 감소하는 값을 표시하기 위해 사용
각 단계는 전체 값의 백분율로 표기된다


- Number 열과 stage열을 dict로 담아 df에 각각 저장
- Concat을 이용해 합친 후 color을 office로 지정해 데이터 구분

```python
df=px.data.election()
fig = px.scatter_ternary(df, a='Joly',b='Coderre',c='Bergeron',
                         hover_name='district',color='winner',size='total',
                         size_max=15,
                         color_discrete_map={'Joly':'blue','Bergeron':'green','Coderre':'red'})
fig.show()
```

#Ternary Plot
정삼각형을 통해 세 변수의 비율을 알아볼 수 있는 시각화 도구
주로 어떤 물질을 구성하는 성분의 비중을 분석하기 위해 물리화학, 암석학, 금속공학 등에서 쓰인다


- px.scatter_ternary()를 통해
각 축에 데이터프레임의 열 선택 후 지정
-hover_name=district로 지정해 팝업 창에 지역구 표기
-size=total로 지정해 total의 크기 순대로 버블 크기 조정

#표본
모집단의 특성을 대표할 수 있는 일부 관측치의 집단
- 여러 개의 특성변수와 목적 변수로 구성
- 표본의 수가 증가하면 표본의 신뢰도 증가
- Random sample : 모집단 내에 크기 n의 표본이 뽑힐 확률이
동일하도록 추출
- 층화추출 : 범주형일 때 범주 내에서 주어진 크기의 random 
sample 추출
-> 머신러닝의 학습/검증/시험데이터 분류
목적변수가 범주형일 때 층화추출 등

#임의성
관측되지 않는 확률변수 -> 오차항
- 목적변수 = 특성변수의 선형/비선형 함수 + 오차항
- 오차항이 존재해야만 모형의 일반화 가능
-> 머신러닝의 일반화 성능 강화(규제화, dropout)

#우도함수
데이터(샘플)이 해당 분포로부터 나왔을 가능도
- 각 샘플에서 후보 분포에 대한 높이를 계산해 다 곱한 것
- (데이터들의 추출은 독립적으로 연달아 일어나기 때문에 곱함)
- 해당 확률분포에서 높은 확률을 가진다-> 데이터를 잘 설명
- 연산의 편의를 위해 우도함수에 log를 씌움
- 로그우도를 최대화 / 음의 로그우도를 최소화 해야 함
-> 머신러닝의 손실함수 대부분이 우도함수로부터 도출

#회귀분석
평가

1. MAE : Mean Absolute Error이며 실제값과 예측값의 차이를 절대값으로 변환해 평균한 것

2. MSE : Mean Squared Error이며 실제값과 예측값의 차이를 제곱해 평균한 것

3. RMSE : MSE 값은 오류의 제곱을 구하므로 실제 오류 평균보다 더 커지는 특성이 있으므로 MSE에 루트를 씌운 것이 RMSE.

4. R^2 : 분산 기반으로 예측 성능을 평가한다. 실제값의 분산 대비 예측값의 분산 비율을 지표로 하며 1에 가까울수록 예측 정확도가 높음
